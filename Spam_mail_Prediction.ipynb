Import Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

Data Preprocessing

```python
#load the dataset to pandas Data Frame
raw_mail_data = pd.read_csv('/content/drive/MyDrive/DataSets/spam.csv', encoding='latin-1', engine ='python')
# replace the null values with a null string
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)), '')
```

```python
mail_data.shape
```

```python
# sample data
mail_data.head()
```

```python
# label spam mail as 0 and non spam mail(ham) as 1
mail_data.loc[mail_data['v1'] == 'spam', 'v1',] = 0
mail_data.loc[mail_data['v1'] == 'ham', 'v1',] = 1
```

```python
# seperate the data as text and label X --> text; Y--> label;
X = mail_data['v2']
Y = mail_data['v1']
```

```python
print(X)
print('''''''''''''''''''')
print(Y)
```

Train Test Split

```python
#split the data as train data and test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.8, test_size=0.2,random_state=3)
```

Feature Extraction

```python
# transform the text data to feature vectors that can be used as input to the SVR model uusing TfidVectorizer
# convert the text to lower case letters
feature_extraction = TfidfVectorizer(min_df=1,stop_words='english',lowercase='True')
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# convert Y_train and Y_test values as Integers
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

Training the model --> Support Vector Machine

```python
# training the support vecotr machine model with training data
model = LinearSVC()
model.fit(X_train_features, Y_train)
```

Evaluation of the model

```python
#prediction on trainiing data
prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

```python
print('Accuracy on training data :' , accuracy_on_training_data)
```

```python
#prediction on test data
prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
```

```python
print('Accuracy on test data :' , accuracy_on_test_data)
```
